```python
url_to_scrape = ('http://web.mta.info/developers/turnstile.html')
r = requests.get(url_to_scrape)
soup = BeautifulSoup(r.text)

sels = '#contentbox > div > div > a'
data = soup.select(sels)
b = []
c = []
for i in range(len(data)):
    b.append(str(data[i]))
    c.append(b[i].partition('"')[-1].rpartition('"')[0])
url_temp = 'http://web.mta.info/developers/{clist}'
url = []
for name in c:
    url.append( url_temp.format(clist=name))
```

```python
main_df = pd.DataFrame()
for numbers in url[30:38]:

    df= pd.read_csv(numbers,
            parse_dates=[['TIME','DATE']], 
            infer_datetime_format= True,
            keep_date_col = True) 
    df.columns = [n.lower().strip() for n in df.columns]
    df['total_entries']= df.groupby(['c/a','unit','scp','station'])['entries'].diff()
    df['total_exits']= df.groupby(['c/a','unit','scp','station',])['exits'].diff()
    df.total_entries.fillna(0, inplace = True)
    df['total_entries'] = abs(df.total_entries)
    df['total_exits'] = abs(df.total_exits)
    df = df [(df.total_entries<=10000) & (df.total_exits<=10000)]
    #df = df[np.abs(df.total_entries-df.total_entries.mean())<=(3*df.total_entries.std())]
    
    if main_df.empty:
        main_df = df
    else:
        
        main_df = main_df.append(df)

main_df['hour'] = main_df.time_date.dt.hour
main_df['day'] = main_df.time_date.dt.dayofweek
main_df['day'].replace([0,1,2,3,4,5,6],['Monday',
                                        'Tuesday',
                                        'Wednesday',
                                        'Thursday',
                                        'Friday',
                                        'Saturday',
                                        'Sunday'],inplace=True)
main_df.head()
```

## Adding data from New Yorker article:

In the following section we have add a new dataset to our MTA dataset in order to make a more accurate 
prediction. This additional data set was originally presented in a New Yorker article called [INEQUALITY AND NEW YORK’S SUBWAY](http://projects.newyorker.com/story/subway/)


```python
df_m= pd.read_csv('income_by_subway_stops.csv')
df_m = df_m.dropna()
df_m.tail()
```

```python
df_mmm = df_m.sort(['income2011'],ascending=False)
```

```python
main_df_ne = main_df
main_df_ne.head()
```

```python
df_mmm = df_mmm.groupby(['stop_name'])[['income2011']].mean().reset_index()
df_mmm['stop_name']  = df_mmm['stop_name'].apply(lambda x: x.upper() if type(x)!=str else x)
```

```python
dt_stations = main_df_ne.groupby(['station'])[['total_entries','total_exits']].mean().reset_index()
dt_stations.head()
```

```python
dt_stations['stop_name'] = dt_stations['station'].apply(lambda x: difflib.get_close_matches(x, df_mmm['stop_name'],n=1,cutoff=.3)[0] if difflib.get_close_matches(x, df_mmm['stop_name'],n=1,cutoff=.3) else x)
```

```python
dt_stations.head()
```

```python
station_list = pd.merge(dt_stations,df_mmm,on='stop_name',how='left')
station_list.head()
```

```python
inc_stations_gb = station_list.groupby(['station'])[['total_entries','total_exits','income2011']].mean()
high_income = inc_stations_gb[inc_stations_gb['income2011']>=100000]
iplot((high_income.sort_values(['total_entries'],ascending=False)).iplot(asFigure=True,
              kind = 'scatter',
              xTitle = 'Station name',
              yTitle = 'Total number of passenger',
              title = 'Total number of passenger'))
```

```python
map_1 = folium.Map(location=[40.7128, -74.0059], zoom_start=12,
                   tiles='Stamen toner')
# add marker clusters

for each in df_mmm.iterrows():
    if each[1]['income2011'] >= 150000:
        ll = [each[1]['lat'],each[1]['long']]
        pp = [each[1]['income2011'], each[1]['stop_name']]
        map_1.polygon_marker(ll, popup=str(pp),fill_color='#0618db',radius=10)
    
    elif each[1].income2011 <= 150000 and each[1].income2011 >= 100000:
                                            ll = [each[1]['lat'],each[1]['long']]
                                            pp = [each[1]['income2011'],each[1]['stop_name']]
                                            map_1.polygon_marker(ll, popup=str(pp),fill_color='#6cdb05',radius=8)

    elif each[1].income2011 <= 100000 and each[1].income2011 >= 50000:
        ll = [each[1]['lat'],each[1]['long']]
        pp = [each[1]['income2011'],each[1]['stop_name']]
        map_1.polygon_marker(ll, popup=str(pp),fill_color='#ffff14',radius=6)
    
    elif each[1].income2011 <= 50000:
         ll = [each[1]['lat'],each[1]['long']]
         pp = [each[1]['income2011'],each[1]['stop_name']]
         map_1.polygon_marker(ll, popup=str(pp),fill_color='#f29113',radius=6)

#map_1.save('nyc.html')
map_1
```
